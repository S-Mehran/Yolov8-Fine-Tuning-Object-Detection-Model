```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")  # Use YOLOv8 Nano for speed; replace with "yolov8m.pt" for more accuracy.
```

```python
# imports pytorch
import torch
# imports the torch_xla package
```

```python
model.train(
    data="/kaggle/input/tt100-4/data.yaml",  # Use absolute path
    epochs=50,
    batch=16,
    imgsz=640,
    lr0=0.01,
    optimizer="Adam",
    workers=4,
    device=0
)
```

```python
# Load the best weights from training
best_model = YOLO('runs/detect/train/weights/best.pt')  # Path to the best weights
```

```python
# Evaluate the fine-tuned model
metrics = model.val(data="/kaggle/input/tt100-4/data.yaml")  # Path to data.yaml
print(metrics)  # Display mAP, precision, recall, etc.
```

```python
import yaml

# Load the dataset config file
data_yaml_path = "/kaggle/input/tt100-4/data.yaml"  # Change if needed

with open(data_yaml_path, "r") as f:
    data_config = yaml.safe_load(f)

# Get validation images path
val_images_path = data_config["val"]
print(f"Validation images are stored in: {val_images_path}")
```

```python
import cv2
import matplotlib.pyplot as plt
import random
import glob

# Get list of validation images
val_images = glob.glob(f"/kaggle/input/tt100-4/{val_images_path}/*.jpg")  # Adjust extension if needed

# Select a few random images
random_images = random.sample(val_images, 3)  # Select 3 images

# Display images
for img_path in random_images:
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis("off")
    plt.title(img_path.split("/")[-1])  # Show filename
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import random
import glob

# Get list of validation images
val_images = glob.glob(f"/kaggle/input/tt100-4/{val_images_path}/*.jpg")  # Adjust extension if needed

# Select a few random images
random_images = random.sample(val_images, 3)  # Select 3 images

# Display images
for img_path in random_images:
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis("off")
    plt.title(img_path.split("/")[-1])  # Show filename
    plt.show()
```

```python
for img_path in random_images:
    results = best_model.predict(img_path, save=True)  # Run detection on image
    
    # Load the image with predictions
    pred_img_path = results[0].save_dir / img_path.split("/")[-1]  # YOLO saves predictions in 'runs/detect/predict'
    
    pred_img = cv2.imread(str(pred_img_path))
    pred_img = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)

    # Show predicted image with bounding boxes
    plt.figure(figsize=(6, 6))
    plt.imshow(pred_img)
    plt.axis("off")
    plt.title("Predicted: " + img_path.split("/")[-1])
    plt.show()
```
